```python
import numpy as np

class Perceptron:
  #
    def __init__(self, input_size, lr=0.1):
        self.weights = np.zeros(input_size + 1) #
        self.lr = lr

    def predict(self, inputs):
        sum = np.dot(inputs, self.weights[1:]) + self.weights[0]
        return 1 if sum > 0 else 0

    def train(self, training_inputs, labels, epochs):
       for epoch in range(epochs):
           correct = 0
           for inputs, label in zip(training_inputs, labels):
               prediction = self.predict(inputs)
               error = label - prediction
               if error == 0:
                    correct += 1
               self.weights[1:] += self.lr * error * inputs
               self.weights[0] += self.lr * error

           accuracy = correct / len(training_inputs)
           print(f"Epoch {epoch+1}/{epochs} - Accuracy: {accuracy}, Weights: {self.weights}")
```

```python
# AND все должны быть единичками
and_inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
and_labels = np.array([0, 0, 0, 1])

# OR одно из должно быть
or_inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
or_labels = np.array([0, 1, 1, 1])

# XOR хотя бы одно, но не оба единички
xor_inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
xor_labels = np.array([0, 1, 1, 0])
```

```python
def evaluate(perceptron, inputs, labels):
    correct = 0
    for i, label in zip(inputs, labels):
        if perceptron.predict(i) == label:
            correct += 1
    return correct / len(inputs)

# AND
and_perceptron = Perceptron(2)
and_perceptron.train(and_inputs, and_labels, epochs=50)
and_accuracy = evaluate(and_perceptron, and_inputs, and_labels)

# OR
or_perceptron = Perceptron(2)
or_perceptron.train(or_inputs, or_labels, epochs=50)
or_accuracy = evaluate(or_perceptron, or_inputs, or_labels)

# XOR
xor_perceptron = Perceptron(2)
xor_perceptron.train(xor_inputs, xor_labels, epochs=50)
xor_accuracy = evaluate(xor_perceptron, xor_inputs, xor_labels)

print(f"AND точность: {and_accuracy}")
print(f"OR точность: {or_accuracy}")
print(f"XOR точность: {xor_accuracy}")
```

```python

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

class MLP:
    def __init__(self, input_size, hidden_size, output_size, lr=1):
        self.w1 = 2 * np.random.random((input_size, hidden_size)) - 1
        self.w2 = 2 * np.random.random((hidden_size, output_size)) - 1
        self.lr = lr

    def train(self, inputs, labels, epochs):
        for epoch in range(epochs):
            l1 = sigmoid(np.dot(inputs, self.w1))
            l2 = sigmoid(np.dot(l1, self.w2))

            l2_error = labels - l2
            l2_delta = l2_error * sigmoid_derivative(l2)

            l1_error = l2_delta.dot(self.w2.T)
            l1_delta = l1_error * sigmoid_derivative(l1)

            self.w2 += l1.T.dot(l2_delta) * self.lr
            self.w1 += inputs.T.dot(l1_delta) * self.lr

    def predict(self, inputs):
        l1 = sigmoid(np.dot(inputs, self.w1))
        l2 = sigmoid(np.dot(l1, self.w2))
        return l2

# XOR
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
labels = np.array([[0], [1], [1], [0]])

mlp = MLP(2, 5, 1, lr=1)
mlp.train(inputs, labels, epochs=20000)

print("Predictions after training")
print(mlp.predict(inputs))
```
